```python
from parser_env import ParserEnv
env = ParserEnv()
```

```python
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import A2C
load_file = "modelv1_params"
model = A2C('MlpPolicy', env, verbose=1)
```

```python
model.set_parameters(load_file, exact_match=False)
```

```python
# model.save("modelv1_params")
model.learn(total_timesteps=250000)
```

```python
scores = []
times = []
for i in range(1000):
  # print("Started episode ", i)
  obs = env.reset()
  score = 0
  t = 0
  done = False
  while not done: 
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    if t > 1000:
      print(f"Episode: {i}, Time Step: {t}, Action: {action}, Observation: {obs}, Reward: {reward}, Done: {done}")
    score += reward
    t += 1
  scores.append(score)
  times.append(t)
  print(f"Episode {i}; Score: {score}, Time: {t} ")
print(f"\nAverage Score: {sum(scores)/len(scores)}, Average Time: {sum(times)/len(times)}")
```
